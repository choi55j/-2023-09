```python
labels = [
    "영어",
    "잠",
    "포토샵",
    "액셀",
    "수학",
    "국어",
    "운동"
]
time=[
    2,
    3,
    3,
    1.5,
    3,
    2,
    2
]
colors = ["darkblue", "#6d597a", "#b56576", "#e56b6f", "#eaac8b","pink","red"]
```
