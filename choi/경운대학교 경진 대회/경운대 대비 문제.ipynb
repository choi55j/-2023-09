```python
import csv
import matplotlib.pyplot as plt

f = open('data/age_final.csv')
data = csv.reader(f)

result = []
name = input('입력하세요')
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.rc('font', family = 'Malgun Gothic')
plt.plot(result, label = '연령별분포', color = 'red')
plt.title(name+' 인구분포')
plt.legend()
plt.show()
```
