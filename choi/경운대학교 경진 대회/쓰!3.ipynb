```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('pop202306.csv')
data = csv.reader(f)

result = []
name = input('지역을 입력하시오 : ')
for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + ' 인구분포')
plt.plot(result,'red',label = '연령별 분포')
plt.legend()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

t = int(input('시간을 입력하면 승차인원아 가장 많은 역을 찾습니다. : '))
print()

mx = 0
mx_station=''

for row in data:
    row[4:] = map(int, row[4:])
    cell_index = row[2*t-4]
    if cell_index > mx:
        mx = cell_index
        mx_station = row[3]+'('+row[1]+') : '
        print(mx_station, mx)
print('--------------------------------')
print()
print(mx_station, mx)
```
