```python
import csv
import matplotlib.pyplot as plt
f = open('temp2013.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1]!='':
        if row[6].split('-')[1] == '08' :
            result.append(float(row[5]))
plt.plot(result, 'r')
plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('temp1904.csv')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data:
    if row[-1] !='' and row[-2] != '':
        if 1904 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '20' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.plot(high,'red', label="high")
plt.plot(low,'blue', label = "low")
plt.legend(loc = 2)
plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('temp1904.csv')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1904 <=int(row[0].split('-')[0]):
            if row[0].split('-')[1] == "10" and row[0].split('-')[2] == "20" :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title("내 성년 이후의 생일의 기온 그래프")
plt.plot(high,'hotpink',label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
plt.show()
f.close()
```
